##丟到Random Forest裡面試試看吧!

```python
import csv as csv
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier 
#修剪訓練資料
train = pd.read_csv('train.csv', header = 0)
train['Gender'] = train['Sex'].map({'female': 0, 'male': 1}).astype(int) 
train['Embark'] = train['Embarked'].map({'C': 1, 'Q': 2, 'S': 3}) #換成數字
train.loc[train['Age'].isnull(), 'Age'] = train['Age'].mean() #用平均填滿
train = train.drop(['PassengerId', 'Name', 'Sex', 'Ticket', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked'], axis = 1)
train = train.dropna() #丟掉不完整、不需要的
print train
train_data = train.values #DataFrame換成陣列
train_data = train_data.astype(np.float) #轉型成float
forest = RandomForestClassifier(n_estimators = 100) #訓練Random Forest
forest = forest.fit(train_data[0::, 1::], train_data[0::,0])
#修剪測試資料
test = pd.read_csv('test.csv', header = 0)
test.loc[test['Age'].isnull(), 'Age'] = train['Age'].mean()
idx = test['PassengerId']
idx = idx.values #存編號
test['Gender'] = test['Sex'].map({'female': 0, 'male': 1}).astype(int)
test['Embark'] = test['Embarked'].map({'C': 1, 'Q': 2, 'S': 3})
test = test.drop(['PassengerId', 'Name', 'Sex', 'Ticket', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked'], axis = 1)
#預測測試資料
test_data = test.values
test_data = test_data.astype(np.float)
output = forest.predict(test_data)
score = forest.score(train_data[0::, 1::], train_data[0::,0])
#寫入結果
pdtFile = open('predict2.csv', 'wb')
pdt = csv.writer(pdtFile)
pdt.writerow(["PassengerId", "Survived"])
for i in range(0, len(output)):
    pdt.writerow([idx[i], output[i]])
    print "result #%d: %d" % (idx[i], output[i])
print score



```
