```python
import requests
from bs4 import BeautifulSoup

#把網頁抓下來
r = requests.get("http://www.technologyreview.com/computing/stream/")
s = BeautifulSoup(r.text)

#找到文章串
stream = s.select("ul.stream")
posts = stream[0].select("li.item")

#建立一個list然後把每篇的標題跟時間做成dict塞進去
result = []
for i in posts:
    title = i.select("h1 > a")[0].get_text()
    date = i.time["datetime"]
    result.append({"title": title, "date": date})

print result
```
